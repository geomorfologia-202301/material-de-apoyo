```python
import geopandas as gpd
import geoplot as gpt
```

Intentemos cargar los datos

```python
mi_data = gpd.read_file('6nLwFTBa.gpkg', )
mi_data.head()
```

```python
mi_data.crs='epsg:32619'
mi_data.crs
```

Intentemos hacer el mapa

```python
mi_data.plot()
mi_data_ll = mi_data.to_crs('EPSG:4326')
gpt.polyplot(mi_data_ll, figsize=(8, 4))
```
